```python
import pandas as pd
base = pd.read_csv('breast-cancer.csv')
base.head()
```

```python
base['mitoses'].unique()
```

```python
def converte_classe(rotulo):
    if rotulo == 'benign':
        return 1
    else:
        return 0
```

```python
def unkown(rotulo):
    if rotulo == '?':
        return 0
    else:
        return rotulo
```

```python
base['class'] = base['class'].apply(converte_classe)
```

```python
base['bare_nuclei'] = base['bare_nuclei'].apply(unkown)
```

```python
base.head()
```

```python
base['class'].unique()
```

```python
X = base.drop(['class', 'code_number'], axis = 1)
```

```python
X.head()
type(X)
```

```python
y = base['class']
```

```python
y.head()
```

```python
type(y)
```

```python
import tensorflow as tf
```

```python
X.columns
```

```python
nome_colunas = ['clump_thickness', 'uniformity_cell_size', 'uniformity_cell_shape',
       'marginal_adhesion', 'single_epitelial_cell_size', 'bare_nuclei',
       'bland_chromation', 'normal_nucleoli', 'mitoses']
```

```python
colunas = [ tf.feature_column.numeric_column(key = c) for c in nome_colunas ]
```

```python
print(colunas)
```

```python
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.3)
```

```python
X_treinamento.shape
type(X_treinamento)
```

```python
X_teste.shape
type(y_treinamento)
```

```python
funcao_treinamento = tf.compat.v1.estimator.inputs.pandas_input_fn(x = X_treinamento, y = y_treinamento,
                                                        batch_size = 32, num_epochs = None, shuffle = True)
```

```python
classificador = tf.estimator.LinearClassifier(feature_columns=colunas)
```

```python
classificador.train(input_fn=funcao_treinamento, steps = 10000)
```

```python
funcao_previsao = tf.compat.v1.estimator.inputs.pandas_input_fn(x = X_teste, batch_size = 32, shuffle = False)
```

```python
previsoes = classificador.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
previsoes_final = []
for p in classificador.predict(input_fn=funcao_previsao):
    previsoes_final.append(p['class_ids'])
```

```python
previsoes_final
```

```python
from sklearn.metrics import accuracy_score
taxa_acerto = accuracy_score(y_teste, previsoes_final)
taxa_acerto
```
